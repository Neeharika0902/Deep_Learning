# Data Augmentation

- Data augmentation is a technique used in deep learning to artificially increase the size and diversity of the training dataset without actually collecting new data. It involves applying various transformations to the existing data, such as rotations, translations, scalings, and flips, to create modified versions of the original dataset. 

# Benefits:
Increased Training Data:
- By generating additional training samples from the existing data, data augmentation effectively increases the size of the training dataset. This can help mitigate the problem of having limited data, which is common in many real-world scenarios.

Improved Generalization:
- Augmented data introduces variations that are likely to be encountered in real-world scenarios. This helps the model to generalize better to new, unseen data by making it more robust to variations and distortions.

Reduced Overfitting:
- Overfitting occurs when a model learns the training data too well, including the noise and details that do not generalize to new data. Data augmentation helps in reducing overfitting by providing a more diverse set of training examples, which forces the model to learn the underlying patterns rather than memorizing the training data.

# Common Data Augmentation Techniques

Geometric Transformations:
- Rotation: Rotating images by a certain degree.
- Translation: Shifting images along the x or y axis.
- Scaling: Zooming in or out of the images.
- Flipping: Horizontally or vertically flipping the images.
- Shearing: Applying a shearing transformation to the images.


Color Space Transformations:
- Brightness Adjustment: Changing the brightness of the images.
- Contrast Adjustment: Altering the contrast of the images.
- Saturation Adjustment: Modifying the saturation of the images.
- Hue Adjustment: Changing the hue of the images.


Noise Injection:
- Adding random noise to the images to make the model more robust to noisy inputs.

Random Erasing:
- Randomly erasing parts of the image to help the model focus on other features.

```python
from keras.preprocessing import image

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dog_img = image.load_img('dog.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(dog_img)
```

```python
data_aug_fun = ImageDataGenerator(
    rotation_range = 30, # angle
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
# converted into array
dog_img = image.img_to_array(dog_img)
```

```python
dog_img.shape
```

```python
# (number_of_input_image, shape of the image)
input_data = dog_img.reshape(1,200,200,3)
```

# flow
- used for single image

# flow_from_directory
- used to perform augmentation on entire directory of images 

# save_to_dir:
- where you want to store augmented images

```python
i = 0

for output in data_aug_fun.flow(input_data,batch_size=1,save_to_dir='Output_Data'):

    i = i + 1

    if i == 10:
        break
```
