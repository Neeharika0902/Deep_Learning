# Two ways to range data. #

(1) Standardization (mean centering): (Xi - means) / standard_deviation

-> firstly, we are doing mean = 0 and standard_deviation = 1

-> range : -1 to 1

-> when to apply: If you know dont know maximum and minimum value of your data you should go with this.

(2) Normalization : (Xi - Xmin) / (Xmax - Xmin)

-> minimun would be 0 & maximum would be 1.

-> range : 0 to 1

-> when to apply: If you know maximum and minimum value of your data you should go with this.

```python
import pandas as pd
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df  = pd.read_csv('Social_Network_Ads.csv')
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
x = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

# Applying scaling #

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(156,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_log = model.fit(x_train_scaled,y_train,validation_data=(x_test_scaled,y_test), epochs=100)
```

# As you can see, accuracy has been increased !! #
